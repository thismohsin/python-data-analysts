```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
mpg_data = pd.read_json('datasets/auto-mpg.json')
```

```python
mpg_data.replace('.', np.nan, inplace=True)
mpg_data.replace('?', np.nan, inplace=True)

mpg_data = mpg_data.dropna()
mpg_data.head()
```

```python
plt.figure(figsize=(12,8))

sns.stripplot(mpg_data['mpg'])
```

The engine cylinders of a high performance car. The cylinder is the power unit of the engine. This is where fuel is burned and converted into mechanical energy that powers the vehicle. The number of cylinders in a typical car could be four, six or eight. 

* Let's check the mpg value according to the cylinder value. so most of the mpg value under 4. That means most of the car has the cylinder value 4. And cylinder 3 and 5 is vey less. Means vey less number of cars has the cylinder 3 and 5

* This function is similar to stripplot(), but the points are adjusted (only along the categorical axis) 

```python
plt.figure(figsize=(12, 8))

sns.stripplot(x='cylinders', y='mpg',
              data=mpg_data, jitter=True)
```

```python
plt.figure(figsize=(12, 8))

sns.stripplot(x='cylinders', y='mpg', hue='origin',
              data=mpg_data, jitter=True)
```

* This gives a better representation of the distribution of values, but it does not scale well to large numbers of observations. This style of plot is sometimes called a “beeswarm”.
* This will give us more details about the distributions

```python
plt.figure(figsize=(12, 8))

sns.swarmplot(x='cylinders', y='mpg',
              data=mpg_data)
```

```python
plt.figure(figsize=(12, 8))

sns.swarmplot(x='cylinders', y='mpg', hue='origin', dodge=True,
              data=mpg_data)
```
