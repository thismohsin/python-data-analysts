* Univariate analysis is the simplest form of analyzing data. “Uni” means “one”, so in other words your data has only one variable. It doesn’t deal with causes or relationships (unlike regression) and it’s major purpose is to describe; it takes data, summarizes that data and finds patterns in the data.
* The data is technical spec of cars

* This dataset is in csv format so we are converting it into json format using online CSV to JSON converter and then loading it in JSON format.

#### Importing json file and convert it into pandas dataframe

```python
weight_data = pd.read_json('datasets/weight.json')

weight_data.sample(10)
```

```python
weight_data.shape
```

```python
weight_data.describe()
```

```python
plt.figure(figsize=(12,8))

sns.distplot(weight_data['Weight'], color='purple')

plt.title('Weights')

plt.ylabel('Count')
```

* After specifying the bins, it is little mote clear about the value and density of the weight

```python
plt.figure(figsize=(12,8))

sns.distplot(weight_data['Weight'], color='purple', 
             bins=70)

plt.title('Weights')

plt.ylabel('density')
```

```python
plt.figure(figsize=(12,8))

sns.distplot(weight_data['Weight'], hist=False, color='purple')

plt.title('Weights')

plt.ylabel('density')
```

* Using rugplot it is little more clear about the weight. Each weight value will have a line at the bottom

```python
plt.figure(figsize=(12,8))

sns.distplot(weight_data['Weight'], kde=False, 
             rug=True, color='purple')

plt.title('Weights')

plt.ylabel('density')
```

```python
plt.figure(figsize=(12,8))

sns.distplot(weight_data['Weight'], hist=False, 
             rug=True, color='purple')

plt.title('Weights')

plt.ylabel('density')
```

* From the rugplot we can see the frequency of the weight also. If you will check the bttom of the plot, it is very easy to understand weights are most similar 2000 to 3000 also 3500 contains some more weights

```python
plt.figure(figsize=(12,8))

sns.rugplot(weight_data['Weight'], height=0.5, 
            axis='x', color='blue')

plt.title('Weights')

plt.ylabel('density')
```

* The kde plot is giving the outline about the weights of the car. Between 2000 and 2500 has the highest density of weghts. Very less number of the cars contains the weights less than 1500 and greater than 5000

```python
plt.figure(figsize=(12,8))

sns.set(color_codes=True)

sns.kdeplot(weight_data['Weight'], shade=True, color='green')

plt.title('Weights')

plt.ylabel('density')
```

## Bivariate

* Here we can check it out the relation between horsepower and mpg, There are two important factors of a car<br>From this plot we will be able to understand the use of jointplot to show the relationship between horsepower and mpg<br>So here horsepower is inversly proposional to mpg

```python
mpg_data = pd.read_json('datasets/auto-mpg.json')

mpg_data.head()
```

```python
mpg_data.replace('.', np.nan, inplace=True)
mpg_data.replace('?', np.nan, inplace=True)

mpg_data = mpg_data.dropna()
mpg_data.head()
```

```python
plt.figure(figsize=(10, 8))

sns.scatterplot(x='horsepower', y='mpg', 
                data=mpg_data)

plt.xlabel('horsepower')
plt.ylabel('mpg')

plt.show()
```

```python
plt.figure(figsize=(10, 8))

sns.scatterplot(x='horsepower', y='mpg', 
                data=mpg_data, hue='cylinders')

plt.xlabel('horsepower')
plt.ylabel('mpg')

plt.show()
```

```python
plt.figure(figsize=(10, 8))

sns.scatterplot(x='horsepower', y='mpg', 
                data=mpg_data, hue='cylinders', 
                style='cylinders')

plt.xlabel('horsepower')
plt.ylabel('mpg')

plt.show()
```

```python
plt.figure(figsize=(10, 8))

sns.scatterplot(x='horsepower', y='mpg', 
                hue='cylinders',
                size='weight',
                sizes=(30, 300), hue_norm=(0, 9),
                data=mpg_data)

plt.xlabel('horsepower')
plt.ylabel('mpg')

plt.show()
```

```python
sns.jointplot(x='horsepower', y='mpg', data=mpg_data)
```

* Here it is little more clear which horsepower value is most. The value between 60-100 has the high frequency

```python
with sns.axes_style('white'):
    sns.jointplot(x='horsepower', y='mpg', 
                  data=mpg_data, kind='hex',
                  color='k')
```

* Using kde plot also will be able to understand the relationship between the horsepower and mpg

```python
sns.jointplot(x='horsepower', y='mpg', 
              data=mpg_data, kind="kde")
```

* After plotting kde and rugplot, we will able to understand horsepower and mpg value's highest frequency. So here most of the car has the horpower between 50-100 and most of the cars mpg is 20-35

```python
f, ax = plt.subplots(figsize=(6, 6))

sns.kdeplot(mpg_data['horsepower'], mpg_data['mpg'], ax=ax)

sns.rugplot(mpg_data['horsepower'], color="red", ax=ax)

sns.rugplot(mpg_data['mpg'], vertical=True, ax=ax)
```
