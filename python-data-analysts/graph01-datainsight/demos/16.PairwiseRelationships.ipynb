```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
mpg_data = pd.read_json('datasets/auto-mpg.json')

mpg_data.head(5)
```

```python
mpg_data.replace('.', np.nan, inplace=True)
mpg_data.replace('?', np.nan, inplace=True)

mpg_data = mpg_data.dropna()
mpg_data.head()
```

```python
mpg_data_subset = mpg_data[['acceleration', 'horsepower', 'weight', 'mpg']]
```

* Lets's plot a pairplot and find the relationship between each column of the car data
* From here we can see some columns has positive relation and some of the them have negative correlation

```python
sns.pairplot(mpg_data_subset)
```

#### Positively correlated

* Let's plot little more clearly, few columns. horsepower and weights are positively correlated also horsepower and displacement also the value for the positively correlated.
* After defining the colour according to the cylinder value, it is showing cylinder values also positively correlated with these three columns

```python
sns.pairplot(mpg_data, 
             vars=['displacement','horsepower','weight'], 
             hue='cylinders')
```

* Let's see some of the examples for the negatively correlated and neutral.
* From this plot it is showing that horsepower-mpg, horsepower-acceleration are negatively correlated

```python
sns.pairplot(mpg_data, height=3, 
             vars=['mpg', 'acceleration', 'horsepower'],
             hue='cylinders')
```

* Let's make sure the correlation through the regression plot

```python
corrmat = mpg_data.corr()
f, ax = plt.subplots(figsize=(12, 8))

sns.heatmap(corrmat, vmax=.8, square=True, annot=True, 
            fmt='.2f', cmap = 'viridis')
plt.show()
```
