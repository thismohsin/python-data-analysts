```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
```

```python
housing_data = pd.read_csv('datasets/housing.csv')
housing_data.dropna(axis=0, inplace=True)

housing_data.sample(10)
```

```python
households_data = housing_data[['ocean_proximity', 'households']]

households_data.sample(5)
```

```python
grouped_data = households_data.groupby('ocean_proximity', as_index=False).sum()

grouped_data.head()
```

```python
grouped_data.columns
```

```python
plt.pie(grouped_data['households'],
        labels=grouped_data['ocean_proximity'],
        autopct='%.1f')

plt.title('Households by location')
plt.show()
```

```python
COLORS = ['lightblue', 'limegreen', 'orange', 
          'gold', 'peachpuff']
```

```python
plt.pie(grouped_data['households'],
        labels=grouped_data['ocean_proximity'],
        autopct='%.1f',
         
        colors=COLORS)

plt.title('Households by location')
plt.axis('equal')
plt.show()
```

* In this bigger size it is very clear about how much percentage each one has

```python
fig, ax = plt.subplots(figsize=(10, 10))

plt.pie(grouped_data['households'],
        labels=grouped_data['ocean_proximity'],
        autopct='%.1f',
         
        colors=COLORS)

plt.title('Households by location', loc='left')
plt.axis('equal')
plt.show()
```

* Once we will give boundary we will be able to understand more clearly

```python
fig, ax = plt.subplots(figsize=(10, 10))

plt.pie(grouped_data['households'],
        labels=grouped_data['ocean_proximity'],
        autopct='%.1f',
         
        colors=COLORS,
        counterclock=False,
        wedgeprops= {'linewidth' : 2,
                     'edgecolor' : "gray" 
                    })

plt.title('Households by location', loc='left')
plt.axis('equal')
plt.show()
```

```python
explode = (0, 0, 0, 0, 0.2)
```

* From this plot it is very clear about the highest percentage

```python
fig, ax = plt.subplots(figsize=(10, 10))

wedges, texts, autotexts = plt.pie(grouped_data['households'],
                                   labels=grouped_data['ocean_proximity'],
                                   autopct='%.1f',
                                   colors=COLORS,

                                   explode=explode)

wedges[4].set(edgecolor='blue', linewidth=2)
plt.show()
```
