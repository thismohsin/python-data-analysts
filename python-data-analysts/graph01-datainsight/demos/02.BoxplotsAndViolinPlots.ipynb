## Boxplot

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
oil_data = pd.read_csv('datasets/MPOB_data1.csv')
oil_data.dropna(axis=0, inplace=True)

oil_data.head(5)
```

```python
selected_oil_data = oil_data[['GroundnutOil ', 'FishOil ']]
```

```python
selected_oil_data.boxplot(figsize=(12, 8))
```

```python
oil_data.boxplot(column=['Lard '],  by=['Year'], figsize=(12, 8))
```

```python
oil_data.boxplot(column=['Lard '],  by=['Month'], figsize=(12, 8), fontsize=14)
```

```python
COLUMNS = oil_data.columns[2:].values

COLUMNS
```

```python
oil_data_array = oil_data.iloc[:,2:].values

oil_data_array
```

* Let's plot different oil prices and check it out the variations and consistancy
* From this plot we will be able to understand Groundnut oil and Sunflower has extreme value other oils are constant on their prices

```python
fig, ax = plt.subplots(figsize=(12, 8))

bp = plt.boxplot(oil_data_array)

plt.xticks(rotation=70)

plt.xticks([1, 2, 3, 4, 5, 6, 7, 8, 9], COLUMNS)

plt.title('Oil Production 2006-2018')

plt.xlabel('Oils')
plt.ylabel('Prices')

plt.show()
```

### Formatting this new boxplot

#### The colors to use for the boxes

```python
COLORS = ['maroon', 'limegreen', 'blue', 
          'gray', 'gold', 'lightblue', 
          'purple', 'maroon', 'aqua']
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

bp = plt.boxplot(oil_data_array, 
                 patch_artist=True,
                 labels=COLUMNS)

for i in range(len(bp['boxes'])):
    
    bp['boxes'][i].set(facecolor=COLORS[i])
    
    bp['caps'][2*i].set(color=COLORS[i])
    
plt.xticks(rotation=70)
plt.title('Oil Production 2006-2018')
plt.xlabel('Oils')
plt.ylabel('Prices')
plt.show()
```

```python
print('Number of boxes: ', len(bp['boxes']))
print('Number of whiskers: ', len(bp['whiskers']))
print('Number of caps: ', len(bp['caps']))
print('Number of medians: ', len(bp['medians']))
print('Number of fliers: ', len(bp['fliers']))
```

## Violin Plot

* A Violin Plot is used to visualise the distribution of the data and its probability density. This chart is a combination of a Box Plot and a Density Plot that is rotated and placed on each side, to show the distribution shape of the data.

* From here we will be able to understand what is the most of the prices for each oil and here it is very clear about the maximum and minimum

```python
fig, ax = plt.subplots(figsize=(12, 8))

vp = plt.violinplot(oil_data_array)

plt.title('Oil Production 2006-2018')

plt.xlabel('Oils')
plt.ylabel('Prices')

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

vp = plt.violinplot(oil_data_array,
                    showmedians=True)

plt.xticks(rotation=70)
plt.xticks([1, 2, 3, 4, 5, 6, 7, 8, 9], COLUMNS)

plt.title('Oil Production 2006-2018')
plt.xlabel('Oils')
plt.ylabel('Prices')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

vp = plt.violinplot(oil_data_array,
                    showmedians=True)

plt.xticks(rotation=70)
plt.xticks([1, 2, 3, 4, 5, 6, 7, 8, 9], COLUMNS)

for i in range(len(vp['bodies'])):
    
    vp['bodies'][i].set(facecolor=COLORS[i]) 
    
plt.title('Oil Production 2006-2018')
plt.xlabel('Oils')
plt.ylabel('Prices')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

vp = plt.violinplot(oil_data_array,
                    showmedians=True,
                    vert=False)

plt.yticks([1, 2, 3, 4, 5, 6, 7, 8, 9], COLUMNS)

for i in range(len(vp['bodies'])):
    
    vp['bodies'][i].set(facecolor=COLORS[i]) 

plt.title('Oil Production 2006-2018')
plt.xlabel('Prices')
plt.ylabel('Oils')
plt.show()
```
