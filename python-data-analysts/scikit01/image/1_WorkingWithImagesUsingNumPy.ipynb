# Open a color image from Data Module

https://www.pexels.com/photo/blue-geeen-and-orange-parrot-37833/

```python
parrot = skimage.io.imread("./images/pexels-parrot.jpg")
```

```python
plt.figure(figsize=(6,6))   

plt.imshow(parrot, interpolation='nearest'); 
```

## Basic information of 2D color Image 

```python
parrot.shape
```

```python
parrot
```

Filter out those pixels whose <b>pixel value</b> is <b>less than 50</b> 

```python
blue_mask = parrot[:, :, 2] < 15

blue_mask
```

Now, Apply <b>blue mask</b> on filtered out pixels of Parrot image by setting the <b>RGB values as [0, 0, 255]</b>

```python
parrot_copy = parrot.copy()

parrot_copy[blue_mask] = [0, 0, 255]

plt.figure(figsize=(8,8))

plt.imshow(parrot_copy) 
```

```python
red_mask = parrot[:, :, 0] < 15

red_mask
```

```python
parrot_copy = parrot.copy()

parrot_copy[red_mask] = [255, 0, 0]

plt.figure(figsize=(8,8))

plt.imshow(parrot_copy) 
```

# Zoom into an image

<b>Clip out</b> a portion from the image by specifing <b>[r1:c1, r2:c2]</b>

```python
parrot_zoom = parrot[900:1100, 450:900]  

plt.imshow(parrot_zoom)
```

* Create subplots by dividing the figure in <b>1 row, 2 columns</b> 
* mention the figuresize (w,h)
* share X and Y axes to all the subplots 
* <b>ravel()</b> returns a <b>1-d array</b> made out of input 
* Plot the <b>interpolation='nearest'</b> figure on the <b>first position (1,1,1)= 1st row, 1st column and 1st           position</b> 
* Plot the <b>interpolation='gaussian'</b> figure on the <b>second position (1,2,1)= 1st row, 2nd column and 1st         position</b> 

```python
fig, axes = plt.subplots(nrows=1, ncols=2, 
                         
                         figsize=(12,12), 
                         
                         sharex=True, sharey=True)

ax = axes.ravel() 

ax[0].set_title("Interpolation = nearest")
ax[0].imshow(parrot_zoom, interpolation='nearest')

ax[1].set_title("Interpolation = gaussian")
ax[1].imshow(parrot_zoom, interpolation='gaussian')  
```
